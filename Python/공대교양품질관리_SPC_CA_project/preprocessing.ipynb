```python
import pandas as pd
```

```python
def preprocessing(file_name,save):
    data_df = pd.read_json(file_name,lines=True)
    data_pp1 = data_df.drop(['_id','마일리지정보','세부정보'], axis=1)
    yonsei = pd.DataFrame()
    for i in range(0, len(data_pp1.index)):
        df1 = pd.DataFrame(data_pp1.iloc[i,:],).transpose()
        df2 = pd.DataFrame(data_df['마일리지정보'][i].items()).set_index(0).transpose()
        df1 = df1.reset_index()
        df2 = df2.reset_index()
        df3 = pd.concat([df1, df2], axis=1)
        yonsei = yonsei.append(df3)
    yonsei['중간값'] = 0
    yonsei['분산'] = 0
    yonsei['평균학년'] = 0
    yonsei['재수강비율'] = 0
    for i in range(0,len(data_df['세부정보'])):
        df1 = pd.DataFrame()
        print(i/len(data_df['세부정보']), "%", '작업중...')
        for j in range(0, len(data_df['세부정보'][i])):
            df2 = pd.DataFrame(data_df['세부정보'][i][j].items()).set_index(0).transpose()
            df1 = df1.append(df2)
            df1.set_index('순위', inplace=True)
            df1['마일리지'] = pd.to_numeric(df1['마일리지'])
            df1['학년'] = pd.to_numeric(df1['학년'])
            for k in range(0,len(df1)):
                if df1.iloc[k,4] == 'Y':
                    df1.iloc[k,4] = 0
                else:
                    df1.iloc[k,4] = 1
            yonsei.iloc[i,-4] = df1.loc[:,'마일리지'].median()
            yonsei.iloc[i,-3] = df1.loc[:,'마일리지'].var()
            yonsei.iloc[i,-2] = df1.loc[:,'학년'].mean()
            yonsei.iloc[i,-1] = df1.loc[:,'초수강여부'].mean()
    yonsei.to_excel(save)
    return yonsei
```
