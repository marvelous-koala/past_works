# 19년 집단, 20년 집단으로 분할하여 통계상의 유의미한 차이가 존재하는지 확인하기 위해 T 검정 진행
# 유의도 0.05로 설정했으며 이에 해당하는 편의점 & 품목을 산출

```python
import pandas as pd
from pandas import Series, DataFrame
```

# 데이터 적제

```python
cat = pd.read_excel('./04_유통데이터(GS리테일)/04_혁신아이디어분야_유통데이터(GS리테일)_데이터정의서 및 문제 데이터.xlsx', 
              sheet_name = '카테고리별 매출비중')
```

```python
# 카테고리별 매출비중
cat
```

```python
# 대구와 서울로 분류
cat_d = cat[cat.PVN == '대구광역시']
cat_s = cat[cat.PVN == '서울특별시']
```

# 카테고리 데이터 비교(대구)
# 코로나 전 후, 판매 품목의 비중에 변화가 있었는지 확인.

```python
# 필요없는 정보 제거
cat_d = cat_d.drop(['PVN','ANTC_ITEM_LCLS_CD', 'BOR'], axis=1)
```

```python
# 동과 상품종류를 인덱스로 설정하여 정렬
cat_d = cat_d.set_index(['ADMD', 'ANTC_ITEM_LCLS_NM'])
```

```python
# nan은 매출이 없었던 / 판매하지 않는 품목으로 가정하여 0으로 채움
cat_d.fillna(0, inplace=True)
```

```python
# 19년 자료와 20년 자료로 분할
cat_d_19 = cat_d.loc[:,'20190201':'20190531']
cat_d_20 = cat_d.loc[:,'20200201':'20200531']
```

```python
cat_d_19
```

```python
cat_d_result = DataFrame(index=cat_d.index, columns = ['2019'], data = cat_d_19.mean(axis=1))
cat_d_result['2020'] = cat_d_20.mean(axis=1)
cat_d_result['차이'] = cat_d_result['2019'] - cat_d_result['2020']
```

```python
cat_d_result.sort_values(by = '차이', ascending=False).head(50)
```

```python
from scipy.stats import ttest_ind
```

```python
cat_d_result['T 검정값'] = ttest_ind(cat_d_19, cat_d_20, axis=1)[0]
cat_d_result['P Value'] = ttest_ind(cat_d_19, cat_d_20, axis=1)[1]
```

```python
대구_카테고리_유의미 = cat_d_result[cat_d_result['P Value'] <= 0.05]
```

```python
대구_카테고리_유의미.sort_values(by = 'P Value')
```

```python
대구_카테고리_유의미.reset_index(level=['ADMD', 'ANTC_ITEM_LCLS_NM'], inplace= True)
```

```python
대구_카테고리_유의미.pivot_table(index='ADMD', aggfunc= 'count', values = 'ANTC_ITEM_LCLS_NM')\
.sort_values(by = 'ANTC_ITEM_LCLS_NM', ascending= False)
```

```python
대구_카테고리_유의미.pivot_table(index='ANTC_ITEM_LCLS_NM', aggfunc= 'count', values = 'ADMD')\
.sort_values(by = 'ADMD', ascending= False)
```

# 동일한 방식으로 서울의 2019년과 2020년 t test 실행 
# 유의도 0.05 수준에서 p value 검정

```python
cat_s = cat_s.drop(['PVN','ANTC_ITEM_LCLS_CD', 'BOR'], axis=1)
cat_s = cat_s.set_index(['ADMD', 'ANTC_ITEM_LCLS_NM'])
cat_s.fillna(0, inplace=True)
```

```python
cat_s_19 = cat_s.loc[:,'20190201':'20190531']
cat_s_20 = cat_s.loc[:,'20200201':'20200531']
```

```python
cat_s_result = DataFrame(index=cat_s.index, columns = ['2019'], data = cat_s_19.mean(axis=1))
cat_s_result['2020'] = cat_s_20.mean(axis=1)
cat_s_result['차이'] = cat_s_result['2019'] - cat_s_result['2020']
```

```python
cat_s_result.sort_values(by = '차이', ascending=False).head(50)
```

```python
cat_s_result['T 검정값'] = ttest_ind(cat_s_19, cat_s_20, axis=1)[0]
cat_s_result['P Value'] = ttest_ind(cat_s_19, cat_s_20, axis=1)[1]
```

```python
서울_카테고리_유의미 = cat_s_result[cat_s_result['P Value'] <= 0.05]
서울_카테고리_유의미.sort_values(by = 'P Value')
서울_카테고리_유의미.reset_index(level=['ADMD', 'ANTC_ITEM_LCLS_NM'], inplace= True)
```

```python
서울_카테고리_유의미.pivot_table(index='ADMD', aggfunc= 'count', values = 'ANTC_ITEM_LCLS_NM')\
.sort_values(by = 'ANTC_ITEM_LCLS_NM', ascending= False)
```

```python
서울_카테고리_유의미.pivot_table(index='ANTC_ITEM_LCLS_NM', aggfunc= 'count', values = 'ADMD')\
.sort_values(by = 'ADMD', ascending= False)
```
