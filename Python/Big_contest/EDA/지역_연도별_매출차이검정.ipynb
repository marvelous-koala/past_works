```python
import pandas as pd
from pandas import Series, DataFrame
from scipy.stats import ttest_ind
```

```python
# 데이터 적재
동별 = pd.read_excel('./04_유통데이터(GS리테일)/04_혁신아이디어분야_유통데이터(GS리테일)_데이터정의서 및 문제 데이터.xlsx', 
              sheet_name = '동별매출지수')
```

```python
동별 = 동별.set_index(['BOR','ADMD'])
```

```python
# 서울/대구로 분할
서울 = 동별[동별.PVN == '서울특별시']
대구 = 동별[동별.PVN == '대구광역시']
```

```python
# 19년과 20년으로 분할

서울19 = 서울.loc[:,'20190201':'20190531']
서울20 = 서울.loc[:,'20200201':'20200531']
대구19 = 대구.loc[:,'20190201':'20190531']
대구20 = 대구.loc[:,'20200201':'20200531']
```

```python
대구19
```

```python
# 결과 dataframe 생성
서울결과 = DataFrame(index=서울.index, columns = ['2019'], data = 서울19.mean(axis=1))
서울결과['2020'] = 서울20.mean(axis=1)
서울결과['차이'] = 서울결과['2019'] - 서울결과['2020']

대구결과 = DataFrame(index=대구.index, columns = ['2019'], data = 대구19.mean(axis=1))
대구결과['2020'] = 대구20.mean(axis=1)
대구결과['차이'] = 대구결과['2019'] - 대구결과['2020']
```

```python
서울결과.sort_values(by = '차이', ascending=False)
```

```python
서울결과['T 검정값'] = ttest_ind(서울19, 서울20, axis=1)[0]
서울결과['P Value'] = ttest_ind(서울19, 서울20, axis=1)[1]

대구결과['T 검정값'] = ttest_ind(대구19, 대구20, axis=1)[0]
대구결과['P Value'] = ttest_ind(대구19, 대구20, axis=1)[1]
```

```python
서울_유의미 = 서울결과[서울결과['P Value'] <= 0.05]
대구_유의미 = 대구결과[대구결과['P Value'] <= 0.05]
```

```python
# 상위에 위치할수록 유의미한 차이를 보임
# 단위는 %, 0.49라는 의미는 매출이 49% 감소하였다는 뜻.
대구_유의미.sort_values(by = 'P Value')
```
