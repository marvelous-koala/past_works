```python
import pandas as pd
import numpy as np
```

```python
shipping = pd.read_excel("2020 빅콘테스트_CJ올리브네트웍스_제공DB.xlsx", sheet_name='1번')
```

```python
dong = list(shipping.HDNG_NM.unique())
```

```python
from datetime import datetime
```

```python
shipping['DL_YMD2'] = pd.to_datetime(shipping['DL_YMD'], format = '%y%m%d')
```

```python
shipping19 = shipping[(shipping.DL_YMD2 <= '2019-05-31')]
shipping20 = shipping[(shipping.DL_YMD2 > '2019-05-31')]
```

```python
shipping19.set_index('DL_YMD2', inplace= True)
shipping20.set_index('DL_YMD2', inplace= True)
```

```python
dong19 = dong.copy()
dong20 = dong.copy()

for i in range(0, len(dong)):
    dong19[i] = shipping19[shipping19.HDNG_NM == dong[i]]

for i in range(0, len(dong)):
    dong20[i] = shipping20[shipping20.HDNG_NM == dong[i]]
    
    # 0에서 68까지 dong[] 69개의 이름으로 정리
```

```python
kinds = shipping.DL_GD_LCLS_NM.unique()
```

```python
shipping19
```

```python
obj19 = [[] for i in range(len(dong)*len(kinds))]
obj20 = [[] for i in range(len(dong)*len(kinds))]

            
for i in range(0, len(dong)):
    for k in range(0, len(kinds)):
        obj20[10*i + k] = dong20[i][dong20[i].DL_GD_LCLS_NM == kinds[k]]
        print('20 작성중',10*i + k,'/689')
```

```python
# 코로나 확진자 자료(전국) 가져오기
corona = pd.read_excel("확진자 현황.xlsx", sheet_name='종합')
```

```python
corona.set_index('date', inplace = True)
```

```python
상관관계 = pd.DataFrame()

for i in range(0, len(dong)*len(kinds)):
    corona['배송량'] = obj20[i].INVC_CONT
    corona.fillna(0, inplace = True)
    temp_corr = corona.corr()
    temp_corr.drop(['new_cases', 'new_deaths', 'new_tests', '검색량'], axis = 1, inplace= True)
    temp_corr.drop(['배송량'], axis = 0, inplace= True)
    temp_corr['범례'] = i
    상관관계 = pd.concat([상관관계, temp_corr])
    print('작성중', i, '/689')
```

```python
def dong_clf(x):
    return dong[x//10]
def type_clf(x):
    return kinds[x%10]
```

```python
상관관계['동'] = 상관관계['범례'].apply(dong_clf)
상관관계['품목'] = 상관관계['범례'].apply(type_clf)
```

```python
상관관계.index
```

```python
확진자_상관관계 = 상관관계[상관관계.index == 'new_cases']
```

```python
확진자_상관관계.sort_values(by='배송량', ascending=False)
```

```python
temp = 확진자_상관관계[(확진자_상관관계.배송량 >= 0.2) | (확진자_상관관계.배송량 <= -0.2)]
```
