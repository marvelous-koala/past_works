```python
카테고리 = pd.read_excel('04_혁신아이디어분야_유통데이터(GS리테일)_데이터정의서 및 문제 데이터.xlsx', 
              sheet_name = '카테고리(행열변환)')
```

```python
header = 카테고리.iloc[0]
카테고리.rename(columns=header)
```

```python
df = 카테고리[1:].copy()
header = 카테고리.iloc[0]
```

```python
df
```

```python
df['ADMD']=df['ADMD'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d')).copy()
```

```python
df20 = df[df.ADMD >= '2020-01-01'].copy()
```

```python
df20.set_index('ADMD', inplace=True)
df20.fillna(0, inplace=True)
```

```python
corona = pd.read_excel("확진자 현황.xlsx", sheet_name='종합')
```

```python
corona.fillna(0, inplace=True)
corona.set_index('date', inplace = True)
```

```python
자료 = pd.concat([corona, df20], axis=1, sort=True)
```

```python
상관관계 = 자료.corr().iloc[4:,:4].copy()
```

```python
상관관계
# 없으면 식사
# 1번은 간식
# 2번은 마실거리
# 3번은 홈/리빙
# 4번은 헬스/뷰티
# 5번은 취미/여가활동
# 6번은 사회활동
# 7번은 임신/육아
```

```python
상관관계['종합점수'] = abs(상관관계.new_cases) + abs(상관관계.new_deaths) + abs(상관관계.new_tests) + abs(상관관계.검색량)
```

```python
상관관계.sort_values(by = '종합점수', ascending=False).head(50)
```

```python
확진자_상관관계 = 상관관계.drop(axis=1, columns=상관관계.columns[[1,2,3,4]])
```

```python
temp = 확진자_상관관계[(확진자_상관관계.new_cases >= 0.2) | (확진자_상관관계.new_cases <= -0.2)]
```

```python
temp.sort_values(by = 'new_cases')
```
