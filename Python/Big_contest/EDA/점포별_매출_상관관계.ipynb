```python
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc, font_manager
from datetime import datetime
from datetime import timedelta
import numpy as np
```

```python
동별매출 = pd.read_excel('04_혁신아이디어분야_유통데이터(GS리테일)_데이터정의서 및 문제 데이터.xlsx', 
              sheet_name = '동별매출(행열변환)')
```

```python
동별매출
```

```python
동별매출['ADMD']=동별매출['ADMD'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d')).copy()
```

```python
동별매출20 = 동별매출[동별매출.ADMD >= '2020-01-01'].copy()
```

```python
동별매출20.set_index('ADMD', inplace=True)
```

```python
동별매출20
```

```python
corona = pd.read_excel("확진자 현황.xlsx", sheet_name='종합')
```

```python
corona.set_index('date', inplace = True)
```

```python
corona.fillna(0, inplace=True)
동별매출20.fillna(0, inplace=True)
```

```python
자료 = pd.concat([corona, 동별매출20], axis=1, sort=True)
```

```python
상관관계 = 자료.corr().iloc[4:,:4].copy()
```

```python
상관관계['종합점수'] = 상관관계.new_cases + 상관관계.new_deaths + 상관관계.new_tests + 상관관계.검색량
```

```python
상관관계.sort_values(by = '종합점수')
```

```python
확진자_상관관계 = 상관관계.drop(axis=1, columns=상관관계.columns[[1,2,3,4]])
```

```python
temp = 확진자_상관관계[(확진자_상관관계.new_cases >= 0.2) | (확진자_상관관계.new_cases <= -0.2)]
```

```python
temp.sort_values(by = 'new_cases')
```
