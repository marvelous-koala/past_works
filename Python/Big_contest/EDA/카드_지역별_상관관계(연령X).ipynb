```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc, font_manager

from datetime import datetime
from datetime import timedelta
```

```python
raw_내 = pd.read_csv('카드사용량.txt', delimiter='\t')
```

```python
# 날짜를 인덱스로 준 관계로... 다시 가져다가 data time으로 변환하고 다시 인덱스로 돌려준다.
raw_내.reset_index(inplace=True)
raw_내['기준일자']=raw_내['기준일자'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d')).copy()
raw_내.set_index(['기준일자'], inplace = True)
```

```python
# 1회 사용금액 칼럼추가 / 코로나 자료와 동일하게 날짜를 인덱스로 설정
raw_내['1회 사용금액'] = raw_내.이용금액 / raw_내.이용건수
```

```python
raw_내 = raw_내[raw_내.연령 >= 60]
```

```python
남자 = raw_내[raw_내.성별 == 'M']
여자 = raw_내[raw_내.성별 == 'F']
```

```python
동코드 = raw_내.행정동코드.unique()
업종명 = raw_내.업종코드.unique()
동명 = raw_내.동명.unique()
```

```python
남자20 = 남자[남자.연도 == 2020]
여자20 = 여자[여자.연도 == 2020]
```

```python
# 동별로 정렬한 내용을 채워넣을 빈 리스트 형성
동별_남 = [[] for i in range(0,len(동코드))]
동별_여 = [[] for i in range(0,len(동코드))]
```

```python
for i in range(0, len(동코드)):
    동별_남[i] = 남자20[남자20.행정동코드 == 동코드[i]]
    print('남_작성중',i,'/46')

for i in range(0, len(동코드)):
    동별_여[i] = 여자20[여자20.행정동코드 == 동코드[i]]
    print('여_작성중',i,'/46')
```

```python
# 업종은 23가지, 동은 47개 이므로 23*47개 리스트 준비
종합_남 = [[] for i in range(0,len(동코드)*len(업종명))]
종합_여 = [[] for i in range(0,len(동코드)*len(업종명))]
```

```python
# 준비된 리스트에 각각 동/업종으로 구분하여 채워넣기

for i in range(0, len(동코드)):
    for k in range(0, len(업종명)):
        종합_남[len(업종명)*i + k] = 동별_남[i][동별_남[i].업종코드 == 업종명[k]]
        종합_남[len(업종명)*i + k] = 종합_남[len(업종명)*i + k].groupby(level=0).mean()
        print('남 작성중',len(업종명)*i + k,'/1080')
        
for i in range(0, len(동코드)):
    for k in range(0, len(업종명)):
        종합_여[len(업종명)*i + k] = 동별_여[i][동별_여[i].업종코드 == 업종명[k]]
        종합_여[len(업종명)*i + k] = 종합_여[len(업종명)*i + k].groupby(level=0).mean()
        print('여 작성중',len(업종명)*i + k,'/1080')
```

```python
# 코로나 확진자 자료(전국) 가져오기
import numpy as np
corona = pd.read_excel("확진자 현황.xlsx", sheet_name='종합')
```

```python
corona.set_index('date', inplace = True)
```

```python
# 남/녀 각 1081개의 분류와 코로나 사망자, 확진자, 검사자 수의 상관관계를 분석하여 데이터 프레임으로 쌓는다.
# 채워넣을 데이터 프레임 만들기(비어있다고 가정)
상관관계 = pd.DataFrame()


for i in range(0, len(동코드)*len(업종명)):
    corona['이용금액'] = 종합_남[i].이용금액
    corona['이용건수'] = 종합_남[i].이용건수
    corona['1회 사용금액'] = 종합_남[i]['1회 사용금액']
    corona.fillna(0, inplace = True)
    temp_corr = corona.corr()
    temp_corr.drop(['new_cases', 'new_deaths', 'new_tests', '검색량'], axis = 0, inplace= True)
    temp_corr.drop(['이용금액', '이용건수', '1회 사용금액'], axis = 1, inplace= True)
    temp_corr['범례'] = i
    상관관계 = pd.concat([상관관계, temp_corr])
    print('남자_작성중', i, '/1080')
    
for i in range(0, len(동코드)*len(업종명)):
    corona['이용금액'] = 종합_여[i].이용금액
    corona['이용건수'] = 종합_여[i].이용건수
    corona['1회 사용금액'] = 종합_여[i]['1회 사용금액']
    corona.fillna(0, inplace = True)
    temp_corr = corona.corr()
    temp_corr.drop(['new_cases', 'new_deaths', 'new_tests', '검색량'], axis = 0, inplace= True)
    temp_corr.drop(['이용금액', '이용건수', '1회 사용금액'], axis = 1, inplace= True)
    temp_corr['범례'] = i + 1081
    상관관계 = pd.concat([상관관계, temp_corr])
    print('여자_작성중', i, '/1080')
```

```python
# 범례에 알아보기 쉽게 작업하기
# 0 ~ 1080은 남자, 1081 ~ 은 여자
# 23 업종마다 동이 바뀌면서 반복되므로 나머지가 0이면 코드 10(숙박), 1 이면 레저... 등등
```

```python
def FM(x):
    if x <= 1080:
        return '남자', x
    else:
        return '여자', x-1081

def kinds_clf(x):
    return 업종코드[x%23]
    
def dong_clf(x):
    if x <= 1080:
        return 동명[x//len(업종명)]
    else:
        return 동명[(x-1081)//len(업종명)]
```

```python
상관관계['종합점수'] = abs(상관관계.new_cases) + abs(상관관계.new_deaths) + abs(상관관계.new_tests) + abs(상관관계.검색량)
상관관계['성별'] = 상관관계['범례'].apply(FM)
상관관계['업종']= 상관관계['범례'].apply(kinds_clf)
상관관계['동'] = 상관관계['범례'].apply(dong_clf)
```

```python
상관관계.fillna(0, inplace=True)
```

```python
확진자_상관관계 = 상관관계.drop(axis=1, columns=상관관계.columns[[1,2,3,4,5]])
```
