```python
#도시, 행정구 column 추가
age_2019 = age_2019.assign(CITY_NM = '', GU_NM = '')
age_2020 = age_2020.assign(CITY_NM = '', GU_NM = '')
```

```python
#행정구를 구별하는 함수
def where_gu(df):
    i = 0
    for i in list(range(len(df))):
        if df['HDONG_NM'][i] in 서울_중구:
            df['GU_NM'][i] = '서울중구'
        elif df['HDONG_NM'][i] in 서울_노원구:
            df['GU_NM'][i] = '노원구'
        elif df['HDONG_NM'][i] in 대구_중구:
            df['GU_NM'][i] = '대구중구'
        elif df['HDONG_NM'][i] in 대구_수성구:
            df['GU_NM'][i] = '수성구'
        i+1
```

```python
#도시를 구별하는 함수
def where_city(df):
    i = 0
    for i in list(range(len(df))):
        if df['HDONG_NM'][i] in 서울특별시:
            df['CITY_NM'][i] = '서울특별시'
        else:
            df['CITY_NM'][i] = '대구광역시'
        i+1
```

```python
#표 만드는 함수
from scipy.stats import ttest_ind

def result_dataframe(df, df2019, df2020):
    df['2019'] = df2019.groupby('행정동')['전체'].mean()
    df['2020'] = df2020.groupby('행정동')['전체'].mean()
    df['평균 차이'] = np.round(df['2019'] - df['2020'], 2)
    df['감소율(%)'] = np.round((df['2020'] / df['2019'] - 1) * 100, 2)
    for i in list(df2019.행정동.unique()):
        df['T검정값'][i] = ttest_ind(df2019[df2019.행정동 == i]['전체'], df2020[df2020.행정동 == i]['전체'])[0]
        df['p값'][i] = ttest_ind(df2019[df2019.행정동 == i]['전체'], df2020[df2020.행정동 == i]['전체'])[1]
```

```python
def result_dataframe2(df, df2019, df2020):
    for i in list(df2019.행정동.unique()):
        df['T검정값'][i] = ttest_ind(df2019[df2019.행정동 == i]['전체'], df2020[df2020.행정동 == i]['전체'])[0]
        df['p값'][i] = ttest_ind(df2019[df2019.행정동 == i]['전체'], df2020[df2020.행정동 == i]['전체'])[1]
```
