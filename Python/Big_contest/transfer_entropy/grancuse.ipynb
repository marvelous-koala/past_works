```python
from statsmodels.tsa.stattools import grangercausalitytests
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_excel('./log.xlsx')
raw_data.set_index('Date', inplace=True)
```

```python
yr19 = raw_data.loc[20190302 : 20191031]
yr20 = raw_data.loc[20200301 : ]
```

```python
name_list = list(raw_data.columns)
name_list
```

```python
res = grangercausalitytests(yr20[['강남구','동대문구']], maxlag=7)
```

```python
result = list(res[7][0].values())
```

```python
result[0][1]
```

```python
gc = pd.DataFrame(columns=name_list, index=name_list)
```

```python
# loc (index, columns)

gc.loc['강동구', :]
```

```python
gc = pd.DataFrame(columns=name_list, index=name_list)
for i in range(0, len(name_list)):
    for k in range(0, len(name_list)):
        pv = 0
        res = grangercausalitytests(yr19[[name_list[i],name_list[k]]], maxlag=7, verbose=0 )
        result = list(res[7][0].values())
        for j in range(0,4):
            if result[j][1] < 0.05:
                pv += 1
        if pv > 3:
            gc.loc[name_list[i], name_list[k]] = 1
        else:
            gc.loc[name_list[i], name_list[k]] = 0
```

```python
gc.to_excel('year19_gc.xlsx')
```
